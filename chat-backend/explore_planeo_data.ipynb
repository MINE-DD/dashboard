## Build QA Test Data

```python
def qa_category_col_templates(column_name: str) -> list[dict]:
    qa_pairs = []
    unique_col_values = list(data[column_name].unique())
    num_unique_vals = len(unique_col_values)
    
    qa_pairs.append({"question": f"How many {column_name}s are there?", "answer":num_unique_vals})
    qa_pairs.append({"question": f"What {column_name}s are included in the data?", "answer": ", ".join(unique_col_values)})

    return qa_pairs

def qa_numeric_col_templates(column_name: str) -> list[dict]:
    qa_pairs = []

    def split_average_from_range(row):
        if isinstance(row, float):
            return row
        else:
            return float(row.split()[0])


    if column_name == "Prevalence":
        data["average_prevalence"] = data["Prevalence"].apply(split_average_from_range)
        column_name = "average_prevalence"

    max_col_value = float(data[column_name].max())
    min_col_value = float(data[column_name].min())
    avg_col_value = float(data[column_name].mean())
    val_frequencies = data[column_name].value_counts(sort=True)#.head(3) # head for top_k

    most_common_val = val_frequencies.index.to_list()[0]
    most_common_count = list(val_frequencies.values)[0]
    answer_most_common = f"The most common value is {most_common_val} with {most_common_count} occurrences"
    
    qa_pairs.append({"question": f"What is the maximum value of {column_name} found in the data?", "answer":max_col_value})
    qa_pairs.append({"question": f"What is the minimum value of {column_name} found in the data?", "answer":min_col_value})
    qa_pairs.append({"question": f"What is the total average value of {column_name} found in the data?", "answer":avg_col_value})
    qa_pairs.append({"question": f"What is the most frequent value of {column_name} found in the data?", "answer":answer_most_common})

    return qa_pairs

def qa_pathogens_complex(verbose: bool = False):
    qa_pairs = []

    # What is the most prevalent Pathogen in <Site_Location>?
    test_locations = [x.split(",")[0].strip() for x in data["Site_Location"].unique()]# ["Nicaragua", "Ghana", "Bangledesh"]
    for loc in test_locations:
        df = data[data["Site_Location"].str.contains(loc)]
        df_counts = df["Pathogen"].value_counts(sort=True, ascending=False)
        max_count = df_counts.iloc[0]
        pathogens_with_max_count = df_counts[df_counts.values == max_count]
        answer = f"The pathogen(s) with the most occurrences in {loc} are {list(pathogens_with_max_count.index)} with {max_count} occurrences."
        qa_pairs.append({"question": f"What is the most prevalent pathogen in {loc}?", "answer":answer})
        if verbose:
            print(pathogens_with_max_count)
            print(answer)

    # What is the distribution per Age_group of pathogen <Pathogen>? Please normalize in percentages
    test_pathogens = list(data["Pathogen"].unique())# ["Campylobacter spp."]
    for pat in test_pathogens:
        df = data[data["Pathogen"].str.contains(pat)]
        df = df["Age_group"].value_counts(normalize=True).mul(100).round(1).astype(str) + '%'
        stringify_parts = [f"for {idx} it occurs in {val}% of the cases" for idx, val in df.items()]
        if len(stringify_parts) > 0:
            result = ', '.join(stringify_parts[:-1]) + ' and ' + stringify_parts[-1] if len(stringify_parts) > 1 else stringify_parts[0]
        else:
            result = '[UNK]'
        answer = f"The presence of pathogen {pat} is: {result}"
        qa_pairs.append({"question": f"What is the distribution per Age_group of pathogen {pat}? Please normalize in percentages", "answer":answer})
        if verbose: 
            print(df)
            print(answer)

    return qa_pairs
```

```python
planeo_data_QA_tests = [
    {"question": "How many records are there?", "answer": len(data)},
]

for col in ["Pathogen", "Age_group", "Syndrome", "Design", "Age_range", "Site_Location"]:
    planeo_data_QA_tests += qa_category_col_templates(col)
```

```python
planeo_data_QA_tests += qa_numeric_col_templates("Prevalence")
```

```python
planeo_data_QA_tests += qa_pathogens_complex()
```

## Evaluate LLM Outputs

```python
# For the following code, START OUR CHAT API SERVER to get the responses here
import requests
from tqdm import tqdm
from datetime import datetime, timezone

def query_llm(session_id:int, question:str) -> str:
    # Get the current UTC datetime with timezone info
    now_utc = datetime.now(timezone.utc)
    iso_timestamp = now_utc.isoformat(timespec='milliseconds').replace('+00:00', 'Z')

    post_url = f"http://127.0.0.1:8000/chat/{session_id}/message"
    payload = {"content": question, "timestamp": iso_timestamp}
    try:
        response = requests.post(post_url, json=payload, timeout=210).json()
    except requests.Timeout:
        response = {'content': 'ERROR: TimeOut Error Occurred'}
    except requests.RequestException as e:
        response = {'content': f'ERROR: Error {e} Occurred'}
    return response.get('content', 'ERROR: Something is wrong...')

for item in tqdm(planeo_data_QA_tests, unit="Questions", total=len(planeo_data_QA_tests)):
    csv_question = f"In my CSV file: {item['question']}"
    llm_answer = query_llm(1, csv_question)
    item["llm_response"] = llm_answer.replace("\n", " #%# ")
```

```python
df = pd.DataFrame(planeo_data_QA_tests)
df.to_csv("outputs/planeo_data_qa_with_answers.tsv", sep="\t")
df
```
